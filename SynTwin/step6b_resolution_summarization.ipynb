```python
def get_resolution(method):
    threshold_dict = {
        'euclidean': 250,
        'manhattan': 340,
        'cosine': 220,
        'gower': 1060
    }
    return threshold_dict[method]
```

```python
import pandas as pd

output_path = 'data/'
percolation_folder = 'Percolation_threshold'
method_list = ['euclidean','manhattan','cosine','gower']

#ran resolution on HPC for range(100,1000,10) to have a rough result 
for method in method_list:
    if method == 'gower':
        df1 = pd.read_csv(output_path+percolation_folder+'/resolution1000_all_'+method+'.csv').drop(columns=['Unnamed: 0'])
        df2 = pd.read_csv(output_path+percolation_folder+'/resolution1000_all_'+method+'1.csv').drop(columns=['Unnamed: 0'])  
        df = pd.concat([df1, df2]).drop_duplicates()
    else:
        df = pd.read_csv(output_path+percolation_folder+'/resolution1000_all_'+method+'.csv').drop(columns=['Unnamed: 0'])
    df
    best_df = df.sort_values(by='num_10members_communities', ascending=False).groupby(['method','random_state']).head(1)
    best_df.sort_values(by='random_state')
    
    resolution = get_resolution(method)
    best_df[best_df['resolution']==resolution]
    
    best_count = best_df.groupby('method')['resolution'].value_counts().sort_values(ascending=False).head(5)
    best_count
```

```python
def get_random_state(method):
    if method == 'gower':
        df1 = pd.read_csv(output_path+percolation_folder+'/resolution1000_all_'+method+'.csv').drop(columns=['Unnamed: 0'])
        df2 = pd.read_csv(output_path+percolation_folder+'/resolution1000_all_'+method+'1.csv').drop(columns=['Unnamed: 0'])  
        df = pd.concat([df1, df2]).drop_duplicates()
    else:
        df = pd.read_csv(output_path+percolation_folder+'/resolution1000_all_'+method+'.csv').drop(columns=['Unnamed: 0'])
    resolution = get_resolution(method)
    best_df = df.sort_values(by='num_10members_communities', ascending=False).groupby(['method','random_state']).head(1)
    random_seeds = best_df[(best_df['resolution']==resolution)]['random_state'] 
    return random_seeds

random_seeds1 = get_random_state('euclidean')
random_seeds2 = get_random_state('manhattan')
random_seeds3 = get_random_state('cosine')
random_seeds4 = get_random_state('gower')
```

```python
pd.merge(pd.merge(pd.merge(random_seeds1, random_seeds2, on='random_state'), random_seeds3, on='random_state'), random_seeds4, on='random_state')
```
